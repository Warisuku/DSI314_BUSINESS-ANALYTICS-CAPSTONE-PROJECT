```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
# Apriori
freq_items = apriori(df.astype(bool), min_support = 0.05, use_colnames = True, verbose = 1)
freq_items.sort_values("support", ascending = False)
```

```python
df_ar = association_rules(freq_items, metric = "confidence", min_threshold = 0.5)
df_ar
```

```python
df_ar = association_rules(freq_items, metric = "lift", min_threshold = 1)
df_ar
```

```python
# Association Rules & Info
df_ar = association_rules(freq_items, metric = "support", min_threshold = 0.05)
df_ar['union'] = [x | y for x, y in zip(df_ar['antecedents'], df_ar['consequents'])]
df_ar
```

```python
df_ar[(df_ar.support > 0.05) & (df_ar.lift > 1)].sort_values("confidence", ascending = False)
```

```python
df_ar["union"] = df_ar["union"].apply(lambda x: ', '.join(list(x))).astype("unicode")
df_ar
```

```python
rules = df_ar.sort_values("confidence", ascending = False)
```

```python
rules['union'] = [x | y for x, y in zip(rules['antecedents'], rules['consequents'])]
rules
```
